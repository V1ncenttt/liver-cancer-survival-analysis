```python
import pandas as pd

data = pd.read_excel('data/complete_df.xlsx')
data_cleaned = data.iloc[:, 15:]
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier

target = data["Death"]
data_cleaned = data_cleaned.dropna()
target_encoded = LabelEncoder().fit_transform(target[data_cleaned.index])

forest = RandomForestClassifier(n_estimators=100, random_state=42)
forest.fit(data_cleaned, target_encoded)

importances = forest.feature_importances_
feature_importance_df = pd.DataFrame({
    'Feature': data_cleaned.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

correlation_matrix = data_cleaned.corr()
important_features = []
for i in range(428):
    important = feature_importance_df.iloc[i]['Feature']
    if important in data_cleaned.columns:
        correlated_features = correlation_matrix.index[correlation_matrix[important].abs() > 0.7]
        correlated_features_list = list(correlated_features)
        to_drop_list = [x for x in correlated_features_list if x in data_cleaned.columns]
        # Optionally add the important feature to the list
        important_features.append(important)
        data_cleaned = data_cleaned.drop(columns=to_drop_list)
```

```python
data_1=data.iloc[:, :]
drop_list=[x for x in data_1.columns if x not in important_features]
data_1=data_1.drop(columns=drop_list)
data_1.head()
```
